#### create and navigate subplots

Subplots are groups of smaller axes that exist together within a single figure. We call plt.subplot and specify three numbers. They refer to the number of rows, number of columns, and the subplot number you want to activate.
Below is an example of four subplots in a 2 x 2 form:

```python
for i in range (1, 5):
   plt.subplot(2, 2, i)
   plt.text(0.5,0.5, str((2, 2, i)), ha='center', fontsize = 10)
   plt.tight_layout()
```

In the subplots above, the text in the plots are a reference to these three numbers. The ordering of the subplot number is row wise from the top left, starting at the number 1. For example, the top left graph is subplot number 1, the top right graph is subplot number 2, the bottom left graph is subplot number 3, and the bottom right graph is subplot number 4.

__Note__ that you will have to specifically call plt.subplot(2, 2, 1) to activate that subplot.

***plt.tight_layout()*** in the above code, cleans up the padding between subplots for a cleaner look.  Without it the subplots would be squished together.

#### change the transparency for lines
Pass in the alpha parameter to your plot. Alpha can be set from 0 to 1, where 0 is completely transparent and 1 is no transparency.

```python
y1 = [1,10,15,20,25,30,35,40,45,50]
y2 = [1,8,13,18,23,28,33,38,43,48]
y3 = [1,6,11,16,21,26,31,36,41,46]
```

The plot below illustrates how the transparency looks like at alpha levels of 0.9, 0.5, and 0.1

```python
plt.plot(x, y1, alpha = 0.9)
plt.plot(x, y2, alpha = 0.1)
plt.plot(x, y3, alpha = 0.5);
```

#### add a legend
1st add labels to each line

```python
plt.plot(x, y1, label='alpha1')
plt.plot(x, y2, label='alpha2')
plt.plot(x, y3, label='alpha3')
plt.legend();
```

Alternatively you can set a list of labels

```python
mylabels = ['alpha1a','alpha2a','alpha3a']
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(labels = mylabels);
```

#### change the size of my legend

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(labels = mylabels,fontsize = 'x-small');
#fontsize can be 'x-small','small','medium','large','x-large', 
```

#### change the location of where my legend appears

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(labels = mylabels,fontsize = 'large', loc = 'lower right');
```

## AXES

#### name the x and y axis labels and change the axis label fontsize
Depending on which axis you want to work with, you call ‘ylabel’ or ‘xlabel’ as shown below. The first item is the name you want for the axis. To set the size of the font, insert the fontsize parameter as shown below.

```python
plt.ylabel('Sales', fontsize = 15) #for y label
plt.xlabel('Days', fontsize = 15); #for x label
```

```python
plt.ylabel('Sales', fontsize = 15) #for y label
plt.xlabel('Days', fontsize = 15); #for x label
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.legend(labels = mylabels,fontsize = 'large', loc = 'lower right');
```

#### set the tick marks

```python
plt.yticks([0,2,4,6,7,10], ['0', '2B', '4B', '8B', '10B'])
plt.xticks([0,1,2,3,4,5], ['0', '1oz', '2oz', '3oz', '4oz']);
```

#### change the scale

## ANNOTATIONS

### add annotations and arrows to the plot
Create annotations and specify the coordinates to annotate to the xy argument. The xytext define the coordinates of label. If we wanted arrows as well, we will need to use arrowprops in order to make the arrows appear.

```python
plt.plot(x,y)
plt.annotate('local max', xy=(2.2,9), xytext=(4,8.9), arrowprops=dict(facecolor='red',shrink=0.05));
```
